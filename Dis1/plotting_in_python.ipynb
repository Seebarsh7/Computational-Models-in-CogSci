```python
import random
import copy
import numpy as np
import matplotlib.pyplot as plt
```

### Line Chart

```python
rng = np.arange(50)
yrs = 1950 + rng

n_marriages = np.random.randint(0, 10, rng.size)

print("Years")
print(yrs)

print("")

print("Number of times you can legally get married in Texas")
print(n_marriages)

```

```python
plt.plot(yrs, rng+n_marriages)

plt.title('Allowed marriages in Texas by year')
plt.ylabel('Number of times a Texan can get married')
plt.xlim(xmin=yrs[0], xmax=yrs[-1])

```

### Two Line Chart in One Graph

```python


texas = np.random.randint(0, 10,rng.size) * np.arange(rng.size)
nebraska = np.random.randint(0, 10,rng.size) * np.arange(rng.size)



plt.plot(yrs, texas, label="Texas")
plt.plot(yrs,nebraska, label="Nebraska")

plt.title('Allowed marriages by year, or whatever')

plt.legend(loc='upper left')
plt.ylabel('N marriages')
plt.xlim(xmin=yrs[0], xmax=yrs[-1])
plt.tight_layout()
```

### Add Grid to Chart

```python


plt.subplot(2, 1, 1)
plt.plot(yrs, texas, '-', lw=2) #lw = line width
plt.title('Texas')


plt.grid(True)

plt.subplot(2, 1, 2)

plt.plot(yrs, nebraska, '-', lw=2)
plt.title('Nebraska')

plt.grid(True)

plt.tight_layout()
```

### Extra for Line Chart

```python

marriages = np.random.randint(0, 10,size=(2,rng.size)) * 10000


fig, ax = plt.subplots(figsize=(5,2))
ax.stackplot(yrs, marriages, labels=["Texas", "Nebraska"])


ax.set_title('Total marriages by year')

ax.legend(loc='upper left')
ax.set_ylabel('N marriages')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
```

### Bar Chart

```python
barWidth = 0.3

#Y-AXIS (BAR-HEIGHT)
bars1 = np.random.randint(5,20,3)
bars2 = np.random.randint(5,20,3)
 
# ERROR
yer1 = np.random.random(3)
yer2 = np.random.rand(3)
 
# X-AXIS
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
# Create blue bars
plt.bar(r1, bars1, width = barWidth, color = '#3500FF', edgecolor = 'black', yerr=yer1, capsize=7, label='Texas')
 
# Create cyan bars
plt.bar(r2, bars2, width = barWidth, color = '#FF8000', edgecolor = 'black', yerr=yer2, capsize=7, label='Nebraska')
 
# general layout
plt.xticks([r + barWidth for r in range(len(bars1))], ['1850-1900', '1901-1950', '1951-2000'])
plt.ylabel('Something')
plt.legend()
plt.show()
```
