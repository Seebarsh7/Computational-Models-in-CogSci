## 4. Matplotlib (Math Plot Library)

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Simple Pyplot (Line): plt.plot

```python
#creating a series of x containing 50 numbers distributed from -1 to 1. 
x = np.linspace(-1, 1, 50) 
```

```python
print(x)
```

```python
#y1 and y2 are functions of x
y1 = 2*x + 1
y2 = x**2
```

```python
plt.figure(num=3, figsize=(10, 5))
plt.plot(x, y2, label='linear line')
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--', label='square line')
```

```python
plt.figure(num=3, figsize=(8, 5),)
plt.plot(x, y2, label='linear line')
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--', label='square line')
#set the axis
plt.xlim((-1, 2)) #range of x, limit x axis to (-1,2)
plt.ylim((-2, 3)) #range of y
plt.xlabel('I am x') #name the x
plt.ylabel('I am y') #name the y


#plot the figure we just plt.plot
plt.show()
```

```python
plt.figure(num=3, figsize=(8, 5),)
plt.plot(x, y2, label='linear line')
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--', label='square line')
#set the axis
plt.xlim((-1, 2)) #range of x, limit x axis to (-1,2)
plt.ylim((-2, 3)) #range of y
plt.xlabel('I am x') #name the x
plt.ylabel('I am y') #name the y

#set the legend
plt.legend(loc='best') #loc:parameter, put the legend at the best location

#plot the figure we just plt.plot
plt.show()
```

### Scatter Plot: plt.scatter

```python
#scatter plot
n = 1024    # data size
X = np.random.normal(0, 1, n) # normal distribution from 0 to 1
Y = np.random.normal(0, 1, n) # normal distribution from 0 to 1
T = np.arctan2(Y,X) # for color value
plt.xlim()
#alpha = transparency
#s = size
plt.scatter(X, Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.show()
```

### Bar Chart: plt.bar()

```python
#Bar chart
n = 12
X = np.arange(n)
```

```python
#what is np.arange
print(X)
```

```python
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

#simple plot
plt.bar(X, +Y1) #plot the Y1 on the positive Y axis
plt.bar(X, -Y2) #plot the Y1 on the positive X axis

plt.xlim(-.5, n)
plt.ylim(-1.25, 1.25)

#add the tag and change color
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

plt.show()
```

### Extra for Bar Chart

```python
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

#simple plot
plt.bar(X, +Y1) #plot the Y1 on the positive Y axis
plt.bar(X, -Y2) #plot the Y1 on the positive X axis

plt.xlim(-.5, n)
plt.ylim(-1.25, 1.25)

#add the tag and change color
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

#extra bonus: how to add tag to bar chart 
#feel free to skip this part :P
for x, y in zip(X, Y1):
    # ha: horizontal alignment
    # va: vertical alignment
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')

for x, y in zip(X, Y2):
    # ha: horizontal alignment
    # va: vertical alignment
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va='top')
plt.show()
```

### How many times you could marry at Texas

```python
rng = np.arange(50)
yrs = 1950 + rng

n_marriages = np.random.randint(0, 10, rng.size)

print("Years")
print(yrs)

print("")

print("Number of times you can legally get married in Texas")
print(n_marriages)
```

```python
marriages = np.random.randint(0, 10,size=(2,rng.size)) * 10000


fig, ax = plt.subplots(figsize=(10,4))
ax.stackplot(yrs, marriages, labels=["Texas", "Nebraska"])


ax.set_title('Total marriages by year')

ax.legend(loc='upper left')
ax.set_ylabel('N marriages')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
```
