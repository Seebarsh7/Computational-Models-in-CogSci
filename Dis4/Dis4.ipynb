## [Gradient Descent](https://en.wikipedia.org/wiki/Gradient_descent)      
* Gradient descent is a first-order iterative optimization algorithm for finding the minimum of a function. 

### Start from gradient      
$$ gradient = \frac{f(x + \delta)-f(x-\delta)}{2\delta}$$

```python
delta = 0.0001
```

```python
def f(x):
    return x**2 + x
x = 1
der = (f(x + delta) - f(x - delta))/(2*delta)
print(der)
```

What if I have a vector x1,x2

```python
X = [1,2]

for x in X:
    der = (f(x + delta) - f(x - delta))/(2*delta)
    print(der)
```

### Gradient Descent: y=(x+5)Â²  
modified from code in Wikepedia.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-15,5)
y = (x+5)**2
plt.plot(x,y)
plt.show()
```

```python
cur_x = 3 # The algorithm starts at x=3
rate = 0.01 # Learning rate
precision = 0.000001 #This tells us when to stop the algorithm

delta = 1 #the thing you wanna subtract from the guess
max_iters = 100 # maximum number of iterations
iters = 0 #iteration counter
df = lambda x: 2*(x+5) #Gradient of our function 
error = [0]
```

```python
while (delta > precision and iters < max_iters):
    prev_x = cur_x 
    cur_x = cur_x - rate * df(prev_x) 
    delta = abs(cur_x - prev_x)
    error = np.append(error, delta)
    iters = iters+1 #iteration count
    
print("The local minimum occurs at", x)
```

```python
plt.plot(error[1:-1])
plt.show()
```
